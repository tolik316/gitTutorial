```python
def mult_gf(x, y):
    r = 0
    for i in range(8):
        if (y & 0x1) == 1:
            r ^= x
        hbit = x & 0x80
        x = x << 1
        if hbit == 0x80:
            x ^= 0x011d
        y >>= 1
    return r

m_tab = [[0 for j in range(256)] for i in range(256)]
for x in range(256):
        for y in range(256):
            m_tab[x][y] = mult_gf(x, y)

def prepoc_padd(message):
    bytes = [int(message[i * 2: (i + 1) * 2], 16) for i in range(len(message) // 2)]
    binary_message = ''.join(bin(byte)[2:].zfill(8) for byte in bytes)

    N = len(binary_message)
    k = (-97 - N) % l
    binary_N = ''
    for i in range(12):
        binary_N += bin((N >> i * 8) & 0xff)[2:].zfill(8)
    padded_message = binary_message + '1' + '0' * k + binary_N
    stats = []
    for b in range(len(padded_message) // l):
        bits = padded_message[b * l:(b + 1) * l]
        stats.append(int(bits, 2))
    return stats

def xorr(stat, v):
    for j in range(c):
        stat[0][j] ^= (j << 4) ^ v
    return stat

def addd(stat, v):
    for j in range(c):
        word = 0
        for i in range(8):
            word += stat[i][j] << i * 8
        result = word + (0x00f0f0f0f0f0f0f3 ^ ((((c - j - 1) << 4) ^ v) << 56))
        for i in range(8):
            stat[i][j] = (result >> i * 8) & 0xff
    return stat

def sub_byte( stat):
    new_stat = [[0 for j in range(c)] for i in range(8)]
    for i in range(8):
        for j in range(c):
            new_stat[i][j] = sboxes[i % 4][stat[i][j]]
    return new_stat

def shift_byte( stat):
    new_stat = [[0 for j in range(c)] for i in range(8)]
    for i in range(7):
        for j in range(c):
            new_stat[i][(j + i) % c] = stat[i][j]
    for j in range(c):
        shift = 7 if l == 512 else 11
        new_stat[7][(j + shift) % c] = stat[7][j]
    return new_stat

def mix_column( stat):
    new_stat = [[0 for j in range(c)] for i in range(8)]
    for col in range(c):
        for row in range(7, -1, -1):
            product = 0
            for b in range(7, -1, -1):
                product ^= m_tab[stat[b][col]][mds[row][b]]
            new_stat[row][col] = product
    return new_stat

def xor_T( word):
    stat = wrd_to_stat(word)
    for v in range(t):
        stat = xorr(stat, v)
        stat = sub_byte(stat)
        stat = shift_byte(stat)
        stat = mix_column(stat)
    return stat_to_wrd(stat)

def add_T( word):
    stat = wrd_to_stat(word)
    for v in range(t):
        stat = addd(stat, v)
        stat = sub_byte(stat)
        stat = shift_byte(stat)
        stat = mix_column(stat)
    return stat_to_wrd(stat)

def wrd_to_stat( word):
    stat = [[0 for j in range(c)] for i in range(8)]
    hex_word = hex(word)[2:].zfill(l // 4)
    bytes = [int(hex_word[i * 2:(i + 1) * 2], 16) for i in range(len(hex_word) // 2)]
    for j in range(c):
        for i in range(8):
            stat[i][j] = bytes[j * 8 + i]
    return stat

def stat_to_wrd(stat):
    word = 0
    for j in range(c):
        for i in range(8):
            word += stat[i][j] << (c * 8 - 1 - j * 8 - i) * 8
    return word

def kupyna_256_hash( message):
    if l == 512:
        hash_value = 1 << 510
    else:
        hash_value = 1 << 1023
    input_stats = prepoc_padd(message)
    for input_stat in input_stats:
        hash_value = xor_T(hash_value ^ input_stat) ^ add_T(input_stat) ^ hash_value
    return (xor_T(hash_value) ^ hash_value) & int('1' * 256, 2)
```
