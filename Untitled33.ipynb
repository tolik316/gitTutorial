```python
import random

def is_Prime(n):
    if n!=int(n):
        return False
    n=int(n)
    if n==0 or n==1 or n==4 or n==6 or n==8 or n==9:
        return False # перші 6 складних чисел
 
    if n==2 or n==3 or n==5 or n==7:
        return True #перші 4 прості числа
    s = 0
    d = n-1
    while d%2==0: # поки остача не буде дор 0
        d >>=1 # здвиг на 1 біт
        s+=1 
    assert(2**s * d == n-1) # перевірка на правдивість
 
    def trial_composite(a): #
        if pow(a, d, n) == 1: # якщо степінь по модулю = 1
            return False # то це складне число
            
        for i in range(s): #
            if pow(a, 2**i * d, n) ==  n-1:     #якщо степінь по модулю = n-1
                return False # складне число
        return True # якщо ніякі умови не виконались- це просте число
          
 
    for i in range(8):
        a = random.randrange(2, n) #створюємо рандомне число а
        if trial_composite(a): #перевіряємо на складність числа
            return False #
            
          

    print('random numb:', a)  #
    return True #повертаємо те, що число просте якщо жодна з умов складносты не виконалась



```
